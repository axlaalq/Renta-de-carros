```python
import numpy as np
import matplotlib.pyplot as plt

# Par√°metros iniciales
costo_auto = 140000  # Precio en pesos MXN
depreciacion_anual = 0.18  # Devaluaci√≥n anual (18%)
renta_semanal = 2700  # Renta semanal en pesos MXN
semanas_anuales = 52  # Semanas en un a√±o

# C√°lculos
ingreso_anual = renta_semanal * semanas_anuales  # Ingreso anual
valor_residual = costo_auto * (1 - depreciacion_anual)  # Valor del auto despu√©s de un a√±o
devaluacion_anual = costo_auto - valor_residual  # Monto de la devaluaci√≥n anual
ganancia_neta_anual = ingreso_anual - devaluacion_anual  # Ganancia neta anual

# Preparar datos para graficar
a√±os = np.arange(1, 6)  # Analizamos hasta 5 a√±os
valores_residuales = costo_auto * (1 - depreciacion_anual) ** a√±os
devaluaciones = costo_auto - valores_residuales
ganancias_netas = ingreso_anual * a√±os - devaluaciones

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(a√±os, valores_residuales, label="Valor del auto", marker="o")
plt.plot(a√±os, ganancias_netas, label="Ganancia neta acumulada", marker="o", color="green")
plt.bar(a√±os, devaluaciones, label="Devaluaci√≥n acumulada", alpha=0.5, color="red")

# Configuraci√≥n de la gr√°fica
plt.title("Ganancias y devaluaci√≥n de un auto rentado para Uber")
plt.xlabel("A√±os")
plt.ylabel("Pesos MXN")
plt.legend()
plt.grid(True)
plt.show()

ganancia_neta_anual, valores_residuales[-1], ganancias_netas[-1]  # Resultado del primer a√±o y el acumulado final
```
